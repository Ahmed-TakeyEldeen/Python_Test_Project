```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'D:\SQL EXCEL\data_jobs.csv')

df
```

```python
job_salary =  df.groupby('job_title_short')['salary_year_avg'].agg('median').sort_values()

job_salary
```

```python
job_salary.plot(kind='barh')
plt.xticks(rotation= 45,ha='right')
plt.xlabel('USD$')
plt.ylabel('')
plt.title('Median Salary by Job Title')
```
