```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

df= pd.read_csv(r'D:\SQL EXCEL\data_jobs.csv')
df
```

```python
df_japan = df[df['job_country']=='Japan'].copy()
```

```python
df_japan
```

```python
df_japan['job_posted_month']=df_japan['job_posted_date'].dt.strftime('%B')
```

```python
df_japan['job_posted_date'] = pd.to_datetime(df_japan['job_posted_date'])
```

```python
df_japan
```

```python
df_japan_pivot = df_japan.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
```

```python
df_japan_pivot.reset_index(inplace=True)
df_japan_pivot['month_no'] = pd.to_datetime(df_japan_pivot['job_posted_month'],format='%B').dt.month
df_japan_pivot['month_no']
```

```python
df_japan_pivot.sort_values('month_no',inplace=True)
df_japan_pivot.set_index('job_posted_month',inplace=True)
```

```python
df_japan_pivot
```

```python
df_japan_pivot.plot(kind='bar')
```

```python
top4= df_japan['job_title_short'].value_counts().head(4).index.to_list()
```

```python
top4
```

```python
df_japan_pivot[top4].plot(kind='line')

plt.title('Japan_monthly_job_posted')
plt.ylabel('number_of_jobs')
plt.xlabel('')
plt.show()
```

```python
df_soft= pd.read_csv('https://lukeb.co/software_csv')
```

```python
df_soft.set_index('job_posted_month',inplace=True)
```

```python
df_merged=  df_japan_pivot.merge(df_soft,on='job_posted_month')
```

```python
df_merged
```

```python
df_merged = df_merged.drop(columns=['index','month_no'])
```

```python
df_merged
```

```python
top5= (
    df_merged
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .index)
```

```python
top5
```

```python
df_merged[top5].plot(kind='line')

plt.title('merged_jobs')
plt.ylabel('number of jobs')
plt.ylim(0,20000)
plt.xlabel('')
plt.xticks(rotation=90)
## plt.legend().remove()
plt.show()
```
