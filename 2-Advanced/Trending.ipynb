```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

df = pd.read_csv(r'D:\SQL EXCEL\data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
df_DA
```

```python
df_DA['job_posted_month'] = df_DA['job_posted_date'].dt.month
df_DA
```

```python
df_DA_Explode = df_DA.explode('job_skills')
df_da_pivot=  df_DA_Explode.pivot_table(index='job_posted_month' ,columns='job_skills' ,aggfunc='size',fill_value=0)
```

```python
df_da_pivot.loc['Total'] = df_da_pivot.sum()

df_da_pivot
```

```python
df_da_pivot= df_da_pivot[df_da_pivot.loc['Total'].sort_values(ascending=False).index]
df_da_pivot= df_da_pivot.drop('Total')
```

```python
df_da_pivot.iloc[: , :5].plot(kind='line')
plt.xlabel('')
plt.ylabel('Count')
plt.title('Trending of jobs')
plt.show()
```
