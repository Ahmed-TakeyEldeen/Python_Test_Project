```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

df = pd.read_csv(r'D:\SQL EXCEL\data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f%%')
plt.title('Work from Home')
plt.ylabel('')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title('Job Titles')
plt.ylabel('')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='barh')
plt.title('Job Titles')
plt.gca().invert_yaxis()
plt.ylabel('')
plt.show()
```

```python
df_plo= df[['job_work_from_home','job_no_degree_mention','job_health_insurance']]
df_plo
```

```python
fig ,ax = plt.subplots(1 , 3)

for i , jobs in enumerate(df_plo):
    ax[i].pie(df[jobs].value_counts(),startangle=90,autopct='%1.1f%%',labels =['False','True'])
    ax[i].set_title(jobs)
    

fig.suptitle('Distrbution of jobs')
fig.tight_layout()
```
