```python
import ast
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'D:\SQL EXCEL\data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
import seaborn as sns
```

```python
df_DA_USA = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

df_DA_USA = df_DA_USA.dropna(subset='salary_year_avg')

df_DA_USA
```

```python
df_DA_USA = df_DA_USA.explode('job_skills')

df_DA_USA_Group = df_DA_USA.groupby('job_skills')['salary_year_avg'].agg(['count','median'])

df_DA_top_pay = df_DA_USA_Group.sort_values(by='median' , ascending=False).head(10)

df_DA_top_skills = df_DA_USA_Group.sort_values(by='count' ,ascending=False).head(10)

df_DA_top_skills = df_DA_top_skills.sort_values(by='median',ascending=False).head(10)
```

```python
fig, ax = plt.subplots(2, 1)
sns.set_theme(style='ticks')

sns.barplot(data=df_DA_top_pay ,x='median',y=df_DA_top_pay.index,ax=ax[0] ,palette='dark:b')
# df_DA_top_pay.plot(kind='barh' , y='median',ax=ax[0] ,legend=False)
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x , pos: f'${int(x/1000)}K'))


sns.barplot(data=df_DA_top_skills,x='median',y=df_DA_top_skills.index,ax=ax[1] , palette='light:b_r')
# df_DA_top_skills[:: -1].plot(kind='barh',y='median',ax=ax[1],legend=False)   #other way to invert axis
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda y , pos: f'${int(y/1000)}K'))
ax[1].set_xlim(0,200000)

fig.tight_layout()
```

```python
df_DA_USA['salary_year_avg'].plot(kind='hist', bins=30,edgecolor = 'black')

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x , pos: f'${int(x/1000)}K'))
plt.title('Distribution of USA Data Analyst Yearly Salaries')
plt.xlabel('Yearly Salary')
plt.ylabel('Number of Jobs')
plt.show()
```
