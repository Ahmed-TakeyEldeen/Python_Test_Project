```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')
```

```python
month = df['job_posted_month'].unique()
```

```python
dict_month=  {months : df[df['job_posted_month'] == months] for months in month}
dict_month['May']
```

```python
df_q1= pd.concat([dict_month['January'],dict_month['February'],dict_month['March']],ignore_index=True)
```

```python
df_q1['job_posted_month'].value_counts().plot(kind='bar')
```

```python
df_q2 = pd.concat([dict_month['April'],dict_month['May'],dict_month['June']],ignore_index=True)
```

```python
df_q2['job_posted_month'].value_counts().plot(kind='bar')
```
