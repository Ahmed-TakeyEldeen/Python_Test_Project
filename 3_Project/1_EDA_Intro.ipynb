```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'D:\SQL EXCEL\data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]
df_US
```

```python
df22= df_US['job_location'].value_counts().head(10).to_frame()

df22.plot(kind='barh',colormap='plasma',legend=False)
ax = plt.gca()
sns.despine()
sns.set_theme(style='ticks')
ax.invert_yaxis()
plt.title('Count of Jobs in USA')
plt.show()
```

```python
dict_columns ={
    'job_work_from_home':'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirment',
    'job_health_insurance': 'Health Insurance Offered'
}

fig , ax =plt.subplots(1 , 3)
fig.set_size_inches((12,5))

for i ,(column,title) in enumerate(dict_columns.items()):
    ax[i].pie(df_US[column].value_counts(),labels = ['False','True'],autopct = '%1.1f%%',startangle=90)
    ax[i].set_title(title)

plt.show()
```

```python
df22= df_US['company_name'].value_counts().head(10).to_frame()

df22.plot(kind='barh',colormap='plasma',legend=False)
ax = plt.gca()
sns.despine()
sns.set_theme(style='ticks')
ax.invert_yaxis()
plt.title('Count of companies in USA')
plt.show()
```

```python
df_via = df_US['job_via'].value_counts().head(10).to_frame()

df_via.plot(kind='barh',colormap='plasma' ,legend=False)
ax= plt.gca()
ax.invert_yaxis()
sns.despine()
plt.title('Job Found in The Sites')
plt.ylabel('')
plt.show()
```

```python
df_US['job_schedule_type'].value_counts().head(3).plot(kind='area',colormap='plasma')
plt.title('Job Schedule Types')
plt.xlabel('')
sns.despine()
plt.show()
```
