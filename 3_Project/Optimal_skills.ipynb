```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'D:\SQL EXCEL\data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')]
df_US= df_US.dropna(subset='salary_year_avg')
df_US
```

```python
df_US_exp= df_US.explode('job_skills')
df_US_exp
```

```python
df_skills= df_US_exp.groupby(by='job_skills')['salary_year_avg'].agg(['count','median']).sort_values(by='count',ascending=False)
df_skills = df_skills.rename(columns={'count': 'skill_count','median': 'median_salary'})
df_len = len(df_US)
df_skills['skill_percent'] = df_skills['skill_count']/df_len * 100

skill_percent = 5

df_high_demand_skills = df_skills[df_skills['skill_percent'] >5]

df_high_demand_skills
```

```python
from adjustText import adjust_text

df_high_demand_skills.plot(kind='scatter', x='skill_percent',y='median_salary')
plt.tight_layout()
sns.despine()
plt.title('Top 10 optimal skills for Data Analyst')
texts=[]

for i , txt in enumerate(df_high_demand_skills.index):
    texts.append(plt.text(df_high_demand_skills['skill_percent'].iloc[i],df_high_demand_skills['median_salary'].iloc[i],txt))

adjust_text(texts,arrowprops=dict(arrowstyle='->',color='gray'))

ax =plt.gca()
ax.xaxis.set_major_formatter(lambda x ,_ : f'{int(x)}%')
ay= plt.gca()
ay.yaxis.set_major_formatter(lambda y ,_ : f'{int(y/1000)}K')
plt.xlabel('Skill Percentage')
plt.ylabel('Median Salary(USD)')
plt.show()
```

```python
df_technology = df['job_type_skills'].copy()

df_technology = df_technology.drop_duplicates()

df_technology = df_technology.dropna()

technology_dict = {}

for row in df_technology:
    row_dict = ast.literal_eval(row)
    for key ,value in row_dict.items():
        if key in technology_dict:
            technology_dict[key] += value
        else:
            technology_dict[key] = value

for key , value in technology_dict.items():
    technology_dict[key] = list(set(value))

technology_dict
```

```python
technology_dict = pd.DataFrame(list(technology_dict.items()),columns=['technology','skills'])
technology_dict
```

```python
technology_dict = technology_dict.explode('skills')
technology_dict
```

```python
df_plot= df_high_demand_skills.merge(technology_dict,left_on='job_skills',right_on='skills')
df_plot
```

```python
from adjustText import adjust_text

#df_high_demand_skills.plot(kind='scatter', x='skill_percent',y='median_salary')
sns.scatterplot(
    data=df_plot,
    x='skill_percent',
    y='median_salary',
    hue='technology'
)
plt.tight_layout()
sns.despine()
plt.title('Top 10 optimal skills for Data Analyst')
texts=[]

for i , txt in enumerate(df_high_demand_skills.index):
    texts.append(plt.text(df_high_demand_skills['skill_percent'].iloc[i],df_high_demand_skills['median_salary'].iloc[i],txt))

adjust_text(texts,arrowprops=dict(arrowstyle='->',color='gray'))

ax =plt.gca()
ax.xaxis.set_major_formatter(lambda x ,_ : f'{int(x)}%')
ay= plt.gca()
ay.yaxis.set_major_formatter(lambda y ,_ : f'{int(y/1000)}K')
plt.xlabel('Skill Percentage')
plt.ylabel('Median Salary(USD)')
plt.show()
```
