```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'D:\SQL EXCEL\data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df_UK = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United Kingdom')].copy()
df_UK
```

```python
df_UK['job_posted_month'] = df['job_posted_date'].dt.month
df_UK_explode=  df_UK.explode('job_skills')
df_UK_explode
```

```python
df_UK_pivot= df_UK_explode.pivot_table(index='job_posted_month' , columns='job_skills',aggfunc='size' ,fill_value=0)
df_UK_pivot.loc['Total'] = df_UK_pivot.sum()
df_UK_pivot= df_UK_pivot[df_UK_pivot.loc['Total'].sort_values(ascending=False).index]
df_UK_pivot= df_UK_pivot.drop('Total')
```

```python
df_UK_total= df_UK.groupby(['job_posted_month']).size()
```

```python
df_UK_total
```

```python
df_UK_perc= df_UK_pivot.div(df_UK_total/100,axis=0)
```

```python
df_UK_perc = df_UK_perc.reset_index()
df_UK_perc['job_posted_month_no'] = df_UK_perc['job_posted_month'].apply(lambda x : pd.to_datetime(x , format='%m').strftime('%b'))
df_UK_perc = df_UK_perc.set_index('job_posted_month_no')
df_UK_perc = df_UK_perc.drop(columns='job_posted_month')

df_UK_perc
```

```python
df_plot= df_UK_perc.iloc[: , :5]

sns.lineplot(data=df_plot ,dashes=False , palette='plasma')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analyst for UK')
plt.ylabel('Likelihood in Job Postings')
plt.xlabel('')
plt.legend().remove()
plt.tight_layout()

from matplotlib.ticker import PercentFormatter
ay= plt.gca()
ay.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2 , df_plot.iloc[-2,i] ,df_plot.columns[i])
```
