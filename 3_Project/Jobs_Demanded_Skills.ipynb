```python
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'D:\SQL EXCEL\data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df_US = df[(df['job_country']== 'United States')].copy()
df_US
```

```python
df_skills = df_US.explode('job_skills')
df_skills
```

```python
df_skills_count = df_skills.groupby(['job_skills','job_title_short']).size()
df_skills_count = df_skills_count.reset_index(name='skill_count')
df_skills_count.sort_values(by='skill_count',ascending=False,inplace=True)

df_skills_count
```

```python
job_title = df_skills_count['job_title_short'].unique().tolist()

job_title = sorted(job_title[:3])

job_title
```

```python
fig , ax = plt.subplots(len(job_title),1)

for i ,job_titles in enumerate(job_title):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_titles].head(5)
    df_plot.plot(kind='barh' ,x='job_skills' ,y='skill_count', ax=ax[i],title =job_titles)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].legend().set_visible(False)

fig.suptitle('Counts of Top Skills in Job Postings',fontsize=15)
plt.tight_layout(h_pad=.5)
plt.show()
```

```python
df_job_title_counts=  df_US['job_title_short'].value_counts().reset_index(name='jobs_total')

df_skill_perc=  pd.merge(df_skills_count , df_job_title_counts,how='left',on='job_title_short')

df_skill_perc['skill_percent'] =  100*df_skill_perc['skill_count'] / df_skill_perc['jobs_total']

df_skill_perc
```

```python
fig , ax = plt.subplots(len(job_title),1)

for i ,job_titles in enumerate(job_title):
    df_plot = df_skill_perc[df_skill_perc['job_title_short'] == job_titles].head(5)
    df_plot.plot(kind='barh' ,x='job_skills' ,y='skill_percent', ax=ax[i],title =job_titles)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].legend().set_visible(False)

fig.suptitle('The Percentage of Skills in Job Postings',fontsize=15)
plt.tight_layout(h_pad=.5)
plt.show()
```

```python
#other way to plot

fig , ax = plt.subplots(len(job_title),1)

sns.set_theme(style='ticks')

for i ,job_titles in enumerate(job_title):
    df_plot = df_skill_perc[df_skill_perc['job_title_short'] == job_titles].head(5)
    sns.barplot(data=df_plot ,x='skill_percent',y='job_skills' ,ax=ax[i] ,palette='dark:b_r' ) 
    ax[i].set_title(job_titles)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].set_xlim(0,80)
    # ax[i].legend().set_visible(False) there is a comment when you do ploting using sns it doesnt take ax[i].legend
    for n , v in enumerate(df_plot['skill_percent']):
        ax[i].text(v + 1, n ,f'{v:.0f}%' ,va='center')
    if i != len(job_title) -1:
        ax[i].set_xticks([])

fig.suptitle('The Percentage of Skills in Job Postings',fontsize=15)
plt.tight_layout(h_pad=.5)
plt.show()
```
